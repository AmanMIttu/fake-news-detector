```python
import pandas as pd
import joblib

# Load processed dataset
data = pd.read_csv("data/processed/processed_news.csv")

# Separate text and label
X = data["text"]
y = data["label"]

# Load feature extractor (TF-IDF)
vectorizer = joblib.load("models/feature_extractor.pkl")

# Transform text into features
X_vec = vectorizer.transform(X)

# Load best model (Random Forest)
model = joblib.load("models/random_forest.pkl")

# Make predictions
preds = model.predict(X_vec)

# Save predictions
results = pd.DataFrame({"actual": y, "predicted": preds})
results.to_csv("results/predictions.csv", index=False)

print("Predictions saved!")
```

```python
import sys
sys.path.append("/content/drive/MyDrive/fake-news-detector/src")

import pandas as pd
import joblib
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("data/processed/processed_news.csv")
X = data["text"]
y = data["label"]

# Load feature extractor
feature_extractor = joblib.load("models/feature_extractor.pkl")
X_vec = feature_extractor.transform_tfidf(X)

# Load best model
model = joblib.load("models/random_forest.pkl")

# Predict
preds = model.predict(X_vec)

# Classification report
print(classification_report(y, preds))

# Confusion matrix
cm = confusion_matrix(y, preds)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap="Blues")
plt.title("Confusion Matrix - Fake News Detection")
plt.savefig("confusion_matrix.png")
plt.show()
```

```python
import sys
sys.path.append("/content/drive/MyDrive/fake-news-detector/src")

import pandas as pd
import joblib
from lime.lime_text import LimeTextExplainer

# Load dataset
data = pd.read_csv("data/processed/processed_news.csv")

# Pick a sample article
sample_text = data["text"].iloc[50]

# Load feature extractor + model
feature_extractor = joblib.load("models/feature_extractor.pkl")
model = joblib.load("models/random_forest.pkl")

# prediction function required by LIME
def predict_proba(texts):
    X_vec = feature_extractor.transform_tfidf(texts)
    return model.predict_proba(X_vec)

# LIME explainer
explainer = LimeTextExplainer(class_names=["Real", "Fake"])

# Generate explanation
exp = explainer.explain_instance(sample_text, predict_proba, num_features=10)

# Save explanation
exp.save_to_file("lime_explanation.html")

print("Explanation generated and saved!")
```
